## ğŸ” **VerificaÃ§Ã£o do Dataset**

Vamos fazer uma anÃ¡lise rÃ¡pida para confirmar as caracterÃ­sticas do dataset:

```python
# ğŸ“š Imports necessÃ¡rios
import pandas as pd
from pathlib import Path
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# ğŸ“‚ ConfiguraÃ§Ã£o dos caminhos
DATA_DIR = Path("../data")  # Ajuste se necessÃ¡rio
TRAIN_CSV = DATA_DIR / "train.csv"

print("ğŸ”§ ConfiguraÃ§Ã£o inicial concluÃ­da!")
print(f"ğŸ“ DiretÃ³rio de dados: {DATA_DIR}")
print(f"ğŸ“„ Arquivo CSV: {TRAIN_CSV}")
print(f"âœ… Arquivo existe: {TRAIN_CSV.exists()}")
```

```python
# ğŸ“Š Carregamento e anÃ¡lise bÃ¡sica do dataset
df = pd.read_csv(TRAIN_CSV)

print("=" * 60)
print("ğŸ“Š INFORMAÃ‡Ã•ES BÃSICAS DO DATASET")
print("=" * 60)
print(f"ğŸ“ˆ NÃºmero total de instÃ¢ncias: {len(df):,}")
print(f"ğŸ“‹ Colunas disponÃ­veis: {list(df.columns)}")
print(f"ğŸ’¾ Tamanho do dataset: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# ğŸ” Visualizar as primeiras linhas
print("\n" + "=" * 60)
print("ğŸ‘ï¸ PRIMEIRAS 5 LINHAS DO DATASET")
print("=" * 60)
display(df.head())
```

```python
# ğŸ·ï¸ AnÃ¡lise das classes (multi-label)
print("=" * 60)
print("ğŸ·ï¸ ANÃLISE DAS CLASSES")
print("=" * 60)

# Extrair todas as classes Ãºnicas
all_labels = sorted({lab for labs in df["labels"].astype(str).str.split() for lab in labs})
print(f"ğŸ¯ Classes encontradas: {all_labels}")
print(f"ğŸ“Š Total de classes: {len(all_labels)}")

# Contar ocorrÃªncias de cada classe
cnt = Counter(l for labs in df["labels"].astype(str).str.split() for l in labs)

print("\n" + "=" * 60)
print("ğŸ“ˆ DISTRIBUIÃ‡ÃƒO DE IMAGENS POR CLASSE")
print("=" * 60)

# Criar DataFrame para melhor visualizaÃ§Ã£o
class_distribution = pd.DataFrame([
    {"Classe": k, "Quantidade": v, "Porcentagem": f"{v/len(df)*100:.1f}%"}
    for k, v in sorted(cnt.items(), key=lambda x: x[1], reverse=True)
])

display(class_distribution)

# EstatÃ­sticas adicionais
print(f"\nğŸ“Š EstatÃ­sticas das classes:")
print(f"   â€¢ Classe mais frequente: {max(cnt, key=cnt.get)} ({max(cnt.values()):,} imagens)")
print(f"   â€¢ Classe menos frequente: {min(cnt, key=cnt.get)} ({min(cnt.values()):,} imagens)")
print(f"   â€¢ MÃ©dia de imagens por classe: {sum(cnt.values())/len(cnt):.0f}")
```

```python
# ğŸ”¢ AnÃ¡lise multi-label (quantos rÃ³tulos por imagem)
print("=" * 60)
print("ğŸ”¢ ANÃLISE MULTI-LABEL")
print("=" * 60)

# Calcular nÃºmero de rÃ³tulos por imagem
df['num_labels'] = df['labels'].astype(str).str.split().apply(len)
label_distribution = df['num_labels'].value_counts().sort_index()

print("ğŸ“Š DistribuiÃ§Ã£o do nÃºmero de rÃ³tulos por imagem:")
for num_labels, count in label_distribution.items():
    percentage = count / len(df) * 100
    print(f"   â€¢ {num_labels} rÃ³tulo(s): {count:,} imagens ({percentage:.1f}%)")

print(f"\nğŸ“ˆ EstatÃ­sticas:")
print(f"   â€¢ MÃ©dia de rÃ³tulos por imagem: {df['num_labels'].mean():.2f}")
print(f"   â€¢ MÃ­nimo de rÃ³tulos por imagem: {df['num_labels'].min()}")
print(f"   â€¢ MÃ¡ximo de rÃ³tulos por imagem: {df['num_labels'].max()}")

# Verificar se hÃ¡ imagens sem rÃ³tulo ou com rÃ³tulos problemÃ¡ticos
empty_labels = df[df['labels'].isna() | (df['labels'].astype(str).str.strip() == '')].shape[0]
print(f"   â€¢ Imagens sem rÃ³tulos: {empty_labels}")
```
